```python
import pandas as pd
import numpy as np
import requests
import time
import json
from pathlib import Path

PROCESSED_DIR = Path("../data/processed")

# Load the master dataframe from notebook 2
df = pd.read_csv(PROCESSED_DIR / "student_life_master.csv")
df["date"] = pd.to_datetime(df["date"])

print(f"Master dataframe loaded: {len(df)} rows")
print(f"Date range: {df['date'].min().date()} to {df['date'].max().date()}")
```

```python
# Check how many unique participants and dates we have
print(f"Total rows: {len(df)}")
print(f"Unique participants: {df['hashed_uid'].nunique()}")
print(f"Unique dates: {df['date'].nunique()}")
print(f"Date range: {df['date'].min().date()} to {df['date'].max().date()}")

# Check how many mood responses we'd expect vs what we have
print(f"\nExpected max (48 participants x 71 days): {48 * 71}")
print(f"Actual rows: {len(df)}")

# Check responses per participant
responses_per_participant = df.groupby("hashed_uid")["mood_score"].count()
print(f"\nMood responses per participant:")
print(responses_per_participant.describe())
print(f"\nParticipants with fewer than 5 responses:")
print(responses_per_participant[responses_per_participant < 5])
```
